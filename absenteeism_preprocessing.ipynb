## Grouping the data

The reasons can be grouped into classes that share similar reasons like 1-14 can be a group and so on

```python
type_1 = reason_columns.loc[:,1:14].max(axis=1)
type_2 = reason_columns.loc[:,15:17].max(axis=1)
type_3 = reason_columns.loc[:,18:21].max(axis=1)
type_4 = reason_columns.loc[:,21:].max(axis=1)
```

```python
type_1
```

```python
data = pd.concat([data,type_1,type_2,type_3,type_4],axis=1)
data
```

```python
data.columns.values
```

```python
coumn_name = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours',' Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
data.columns = coumn_name
data.head()
```

```python
column_name_reordered = [' Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data = data[column_name_reordered]
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format = '%d/%m/%Y')
data.head()
```

```python
list_months = []
list_months
```

```python
for i in data.Date:
    list_months.append(i.month)
```

```python
list_months
```

```python
data['Month Value'] = list_months
data.head()
```

```python
day = []

for i in data.Date:
    day.append(i.weekday())
```

```python
data['Day of the week'] = day
```

```python
data.head()
```

```python
data['Education'] = data['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
data['Education'].unique()
```
